## 影响

```python
impact = pd.read_csv("D:/Melody/WeChat/WeChat Files/wxid_4e194sm4yv2822/FileStorage/File/2025-03/影响-已分类-所有-对象国 - 已删未提及.csv")
impact['research countries'] = impact['research countries'].str.split(', ')
im = impact.explode('research countries').reset_index(drop=True)
im
```

```python
countries_i = im.groupby(['research countries']).sum().reset_index()
countries_i
```

```python
filtered_i = countries_i[countries_i['research countries'].isin(countries)]
filtered_i
```

```python
df_melt = filtered_i.melt(id_vars=['research countries', 'title', 'abstract', 'author Country'], 
                  var_name='category', value_name='count')
condition = df_melt['category'] != 'others'
df_melt = df_melt[condition]
category_count = df_melt.groupby(['research countries', 'category'])['count'].sum().reset_index()

top10 = category_count.groupby('research countries').apply(lambda x: x.nlargest(10, 'count')).reset_index(drop=True)
top10
```

```python
# 获取所有不同的类别
all_categories = top10['category'].unique()
num_categories = len(all_categories)
# 选择一个色带
cmap = plt.cm.tab20c
category_colors = {category: cmap(i / num_categories) for i, category in enumerate(all_categories)}
```

```python
from matplotlib.patches import Patch
fig, axes = plt.subplots(2, 5, figsize=(20,5))
c = [['India', 'China', 'Nigeria', 'Brazil', 'United States'], ['Pakistan', 'Mexico', 'Ethiopia', 'Indonesia', 'Iran']]
for i in range(2):
    for j in range(5):
        data = top10[top10['research countries']==c[i][j]]
        bars = sns.barplot(data=data, x='count', y='category', ax=axes[i, j])
        axes[i, j].set_title(c[i][j])
        axes[i, j].tick_params(axis='y', labelsize=7)
        axes[i, j].set_xlabel('')
        axes[i, j].set_ylabel('')
        
        for bar, category in zip(bars.patches, data['category']):
            bar.set_color(category_colors[category])
            

# 创建图例
legend_elements = [Patch(facecolor=category_colors[category], label=category) for category in all_categories]
fig.legend(bbox_to_anchor=(1.05, 0.8), handles=legend_elements, loc='upper left')

plt.tight_layout()
fig.suptitle('Impact', fontsize=25, fontweight='heavy', y=1.1)
plt.savefig('D:/Melody/climate_change/countrie_i.png', bbox_inches='tight')
```
